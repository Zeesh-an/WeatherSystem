```python
import pandas as pd
import numpy as np
from datetime import timedelta, datetime
```

```python
np.random.seed(42)
# Generate dummy data for two days (48 hours)
start_time = pd.to_datetime('2023-12-21 00:00:00')
end_time = start_time + pd.Timedelta(days=2, hours=23)

time_range = pd.date_range(start=start_time, end=end_time, freq='H')

# Generate random data for temperature, humidity, and CO
data = {
    'Time': time_range,
    'Temperature (C)': np.random.uniform(18, 25, len(time_range)),
    'Humidity (%)': np.random.uniform(40, 70, len(time_range)),
    'CO (ppm)': np.random.uniform(0, 5, len(time_range))
}

# Create a DataFrame
df = pd.DataFrame(data)
```
